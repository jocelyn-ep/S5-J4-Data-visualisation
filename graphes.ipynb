```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

dfBeautyProducts = pd.read_csv('company_sales_data.csv', on_bad_lines='skip', delimiter=",")
```

**Graphe 1**

R√©aliser un graphique avec deux axes. L'axe des abscisses va de 0 √† 50 et s'intitule "Axe des x". L'axe des ordonn√©es s'intitule "Axe des y".  
Tracer la droite d'√©quation : x = 3y. Avec x allant de 0 √† 50.  
Le graphe poss√®de le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
#cr√©ation des array x et y avec la fonction linspace
x = np.linspace(0,50,51)
y = x / 3 

#LINEPLOT : relation lin√©aire entre variables continues
plt.plot(x, y, color="r")   
plt.xlabel("Axe des x")
plt.ylabel("Axe des y")
plt.title("Tracer une droite")
plt.show()
```

**Graphe 2**

En premier lieu, nous souhaitons un graphe qui donne la courbe du profit total r√©alis√© chaque mois.

```python
#Courbe avec plt.plot
df = dfBeautyProducts

fig, ax = plt.subplots()
x = df['month_number']
y = df['total_profit']
ax.plot(x, y)   
plt.xlabel("Month")
plt.ylabel("Total profit (‚Ç¨)")
plt.title("Monthly total profit")
# d√©finir les positions et les √©tiquettes des graduations de l'axe x
plt.xticks(range(min(x), max(x)+1, 1))

#Conertir 400000 en 400 000
# Cr√©er la fonction de formatage
def format_y_axis(y, _):
    # Utiliser la m√©thode `format` pour ins√©rer un espace tous les 3 chiffres
    return f"{y:,.0f}".replace(",", " ")
# Utiliser `FuncFormatter` pour appliquer la fonction de formatage √† l'axe y
formatter = mtick.FuncFormatter(format_y_axis)
ax.yaxis.set_major_formatter(formatter)

plt.show()
```

**Graphe 6**

Reprend donc les 3 fichiers csv qui donnent les ventes, la liste des clients et la liste des produits.

D'abord, nous souhaitons avoir un graphe pie chart qui donne la r√©partition des produits par cat√©gorie. Merci d'indiquer le pourcentage par produit (nb de produits de telle cat√©gorie / nb de produits totaux).

```python
#import et nettoyage table customers
dfCustomers = pd.read_csv('customers.csv', on_bad_lines='skip', delimiter=",")
toDrop = dfCustomers.iloc[[2735,8494],:]
dfCustomers = dfCustomers.drop(toDrop.index)

#import et nettoyage table products
dfProducts = pd.read_csv('products.csv', on_bad_lines='skip', delimiter=",")
toDrop = dfProducts.iloc[[731],:]
dfProducts = dfProducts.drop(toDrop.index)

#import et nettoyage table transactions
dfTransactions = pd.read_csv('transactions.csv', on_bad_lines='skip', delimiter=",",)
dfTransactions = dfTransactions.drop_duplicates()

#cr√©ation d'un grand df des ventes avec les jointures
dfSales = dfTransactions.merge(dfCustomers)
dfSales = dfSales.merge(dfProducts)
```

```python
#pie chart avec plt.pie
prodRepByCat = dfProducts.groupby(['categ']).size()

plt.pie(prodRepByCat, labels=["category 0", "category 1", "category 2"],autopct="%1.0f%%")
plt.title("Product repartition among catalogue", fontsize=15)
plt.show()
```
